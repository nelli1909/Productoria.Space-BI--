```python
data.groupby('Бизнес-регион')['Статус клиента (Клиенты)'].value_counts()
```

```python
status_counts = data.groupby(['Бизнес-регион', 'Статус клиента (Клиенты)']).size().unstack(fill_value=0)
```

```python
plt.figure(figsize=(14, 8))
status_counts.plot(kind='bar', stacked=True, colormap='tab20', figsize=(14, 8))

plt.title('Распределение статусов клиентов по регионам')
plt.xlabel('Бизнес-регион')
plt.ylabel('Количество клиентов')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Статус клиента', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
median_infrastructure = data.groupby(['Бизнес-регион', 'Статус клиента (Клиенты)'])['Оценка инфраструктуры (справочно)'].median().unstack()

# Plotting the data as a bar chart
median_infrastructure.plot(kind='bar', stacked=False, figsize=(14, 8), colormap='tab20')

plt.title('Медианная оценка инфраструктуры по регионам и статусам клиентов')
plt.xlabel('Бизнес-регион')
plt.ylabel('Медианная оценка инфраструктуры')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Статус клиента', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
